```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Membaca Gambar
img = Image.open("jokowi-1.jpg")

# ubah ke grayscale
img = img.convert("L")

# ubah menjadi array
img_array = np.array(img)

height, width = img_array.shape
```

```python
# 2. Membuat Kernel Rataan 5x5
kernel = np.ones((5,5)) / 25
```

```python
# 3. Membuat Array Hasil
output = np.zeros((height, width))
```

```python
# 4. Proses Konvolusi Manual
for i in range(2, height-2):
    for j in range(2, width-2):

        total = 0

        for m in range(-2,3):
            for n in range(-2,3):

                pixel = img_array[i+m][j+n]
                weight = kernel[m+2][n+2]

                total += pixel * weight

        output[i][j] = total
```

```python
# 5. Normalisasi Nilai Pixel
output = np.clip(output, 0, 255)
output = output.astype(np.uint8)
```

```python
# 6. Menampilkan Gambar
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Gambar Asli")
plt.imshow(img_array, cmap='gray')
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Hasil Konvolusi 5x5")
plt.imshow(output, cmap='gray')
plt.axis("off")

plt.show()
```

```python
# 7. Menyimpan Hasil Gambar
hasil = Image.fromarray(output)
hasil.save("hasil_konvolusi.jpg")

print("Proses konvolusi selesai!")
print("Hasil gambar disimpan dengan nama: hasil_konvolusi.jpg")
```
