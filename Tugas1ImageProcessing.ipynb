```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('gambarjokowi.jpeg', 0)
```

```python
# Perkalian dengan skalar
# -----------------------------
c = 1.5
hasil_kali = img * c
hasil_kali = np.clip(hasil_kali, 0, 255).astype(np.uint8)
```

```python
# Pembagian dengan skalar
# -----------------------------
c = 2
hasil_bagi = img / c
hasil_bagi = np.clip(hasil_bagi, 0, 255).astype(np.uint8)
```

```python
# Menampilkan hasil (Colab)
# -----------------------------
print("Citra Asli")
cv2_imshow(img)

print("Perkalian Skalar")
cv2_imshow(hasil_kali)

print("Pembagian Skalar")
cv2_imshow(hasil_bagi)
```

```python
import matplotlib.pyplot as plt

plt.imshow(hasil_kali, cmap='gray')
plt.title("Perkalian Skalar")
plt.axis('off')
plt.show()
```

```python
# Menampilkan dalam satu figure
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title("Citra Asli")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(hasil_kali, cmap='gray')
plt.title("Perkalian Skalar")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(hasil_bagi, cmap='gray')
plt.title("Pembagian Skalar")
plt.axis('off')

plt.show()
```
